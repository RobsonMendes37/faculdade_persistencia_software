```python
#LISTA 1
import pandas as pd

df = pd.read_csv("vendas.csv")
df["Total_vendas"] = df['Quantidade'] * df['Preco_Unitario']

Total_vendas = df.groupby('Produto')["Total_vendas"].sum()
# Total_vendas
```

```python

df["Data"] = pd.to_datetime(df["Data"], format='%d/%m/%Y')

vendas_janeiro = df[(df["Data"].dt.month == 1) & (df["Data"].dt.year == 2023)]
# vendas_janeiro

vendas_janeiro.to_csv('vendas_janeiro.csv')
```

```python

import matplotlib.pyplot as plt

# gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(Total_vendas.index, Total_vendas.values)
plt.xlabel("Produto")
plt.ylabel("Vendas Totais")
plt.title("Vendas Totais por Produto")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Ajusta o layout para evitar sobreposição de rótulos
plt.show()
```
